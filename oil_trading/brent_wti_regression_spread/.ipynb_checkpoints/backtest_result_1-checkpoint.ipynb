```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import sys
from regression_signal import rolling_regression_trading_signal, rolling_regression_trading_rule
sys.path.append('../..')
from oil_trading.EquityPerformance import equity_performance
plt.style.use("classic")
%matplotlib inline

path = "https://github.com/queiyanglim/trading_algorithm/blob/master/oil_trading/data/daily.csv?raw=true"
df_pull = pd.read_csv(path, header=[0,1], index_col = 0)

# Prepare data
brent = df_pull.brent.CLOSE
brent.name = "brent"
wti = df_pull.wti.CLOSE
wti.name = "wti"

df_raw= pd.concat([brent, wti], axis = 1)
df = df_raw.copy()
df = df.dropna()

df.head()
```

```python
#load signals
df_signal = rolling_regression_trading_signal(df_input=df, 
                                              X_name= "wti", 
                                              y_name = "brent", 
                                              update_window=20,
                                              ewm_span=170)
df_signal
```

```python

# Create trading rule based on signal
def trend_trading_rule(df_with_signal):
    df = df_with_signal.copy()
    position = []
    pos_cache = 0 # cache position info before appending
    for i, data in df.iterrows():
        if data.spread > data.signal:
            pos_cache = 1 # If spread is widening, follow trend
        elif data.spread < data.signal:
            pos_cache = -1 # If spread is contracting, follow trende
        else:
            pos_cache = 0
        position.append({"timestamp": data.name, "position": pos_cache})
    df_pos = pd.DataFrame(position).set_index("timestamp")
    return pd.concat([df, df_pos], axis=1)

# Refresh timestamp format
# df_signal.index = pd.to_datetime(df_signal.index, format = "%Y-%m-%d")
df_signal[["spread", "fair_spread", "signal"]].plot(figsize=(20,3), grid = True)
plt.show()

# Sample strategy
# trading_strategy = rolling_regression_trading_rule
trading_strategy = trend_trading_rule

# Create position base on trading rule
df_pos = trading_strategy(df_signal)

# Equity performance
df_equity, performance = equity_performance(df_pos, X_name = "wti", y_name = "brent")
df_equity.equity.plot(figsize = (20,3), grid= True, title= "Equity Chart", color ="red")
plt.show()

df_pos[["spread", "fair_spread", "signal"]].iloc[-300:].plot(figsize=(20,3), grid = True)
plt.show()
df_pos["position"].iloc[-300:].plot(figsize=(20,3),grid = True, style = "o")
plt.show()

print(performance)
df_equity
```
